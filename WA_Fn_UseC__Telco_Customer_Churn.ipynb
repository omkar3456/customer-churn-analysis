```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('/content/WA_Fn-UseC_-Telco-Customer-Churn.csv')
df.head(10)
```

```python
df.info()          # shows column types
df.isnull().sum()  # shows missing values
df.shape           # shows rows and columns
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df.dropna(inplace=True)
df['SeniorCitizen'] = df['SeniorCitizen'].map({0: 'No', 1: 'Yes'})
```

```python
df['Churn'].value_counts(normalize=True) * 100
```

```python
sns.countplot(data=df, x='Contract', hue='Churn')
plt.title('Churn by Contract Type')
plt.savefig('churn_by_contract.png')
plt.show()
```

```python
sns.boxplot(data=df, x='Churn', y='MonthlyCharges')
plt.title('Monthly Charges vs Churn')
plt.savefig('monthly_charges_churn.png')
plt.show()
```

```python
sns.histplot(data=df, x='tenure', hue='Churn', bins=30)
plt.title('Tenure vs Churn')
plt.savefig('tenure_churn.png')
plt.show()
```

```python
sns.countplot(data=df, x='InternetService', hue='Churn')
plt.title('Churn by Internet Service')
plt.show()
```

```python
sns.countplot(data=df, x='PaymentMethod', hue='Churn')
plt.xticks(rotation=45)
plt.title('Churn by Payment Method')
plt.tight_layout()
plt.show()
```

```python
sns.countplot(data=df, x='SeniorCitizen', hue='Churn')
plt.title('Churn by Senior Citizen')
plt.show()
```

```python
print("===== PROJECT SUMMARY =====")

total = len(df)
churned = df[df['Churn']=='Yes'].shape[0]
not_churned = df[df['Churn']=='No'].shape[0]

print(f"Total Customers: {total}")
print(f"Customers Who Left: {churned}")
print(f"Customers Who Stayed: {not_churned}")
print(f"Churn Rate: {round(churned/total*100, 2)}%")

print("\n--- By Contract Type ---")
print(df.groupby('Contract')['Churn'].value_counts(normalize=True).round(2)*100)

print("\n--- Average Monthly Charges ---")
print(df.groupby('Churn')['MonthlyCharges'].mean().round(2))

print("\n--- Average Tenure ---")
print(df.groupby('Churn')['tenure'].mean().round(2))
```
